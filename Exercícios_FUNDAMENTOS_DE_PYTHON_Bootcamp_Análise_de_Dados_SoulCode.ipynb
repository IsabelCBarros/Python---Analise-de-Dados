2 - Crie um script que aproveite a lista utilizada para imprimir a mensagem do exerccio
anterior e utilize uma estrutura for que percorra toda a lista e faa a tipagem de cada
valor como inteiro, aps isso acesse o ndice 0 dessa nova lista e verifique seu tipo e
caso seja o tipo inteiro imprima para o usurio a seguinte mensagem: este valor  do
tipo inteiro, caso contrrio imprima: este valor no  do tipo inteiro.

```python
dados = input('Informe uma sequncia de nmeros inteiros separados por espaos: ')
dados = dados.split()
dados_tipados = []
for i in range(len(dados)):
    dados_tipados.append(dados[i])
    dados_tipados[i] = int(dados_tipados[i])
if type(dados_tipados[0]) == int:
    print(f'Este valor {dados_tipados[0]}  do tipo inteiro')
else:
    print(f'Este valor {dados_tipados[0]} no  do tipo inteiro')
```

4 - Crie uma lista vazia denominada quadrados e utilizando a lista tipada como int do
exerccio 2, faa uma estrutura do tipo for que percorra cada elemento dessa lista de
nmeros inteiros, eleve cada valor encontrado ao quadrado e insira cada resultado na
lista quadrados e ao final exiba a soma dos valores da lista quadrados.

```python
dados = input('Informe uma sequncia de nmeros inteiros separados por espaos: ')
dados = dados.split()
dados_tipados = []
quadrados = []
for i in range(len(dados)):
    dados_tipados.append(dados[i])
    dados_tipados[i] = int(dados_tipados[i])
    quadrados.append(dados_tipados[i]**2)
print(f'a soma de todos os valores informados elevado ao quadrado  {sum(quadrados)}')
```

5 - Crie um programa que receba do usurio duas listas de n nmeros inteiros digitados
pelo usurio separados por espaos e aps isso utilize o mtodo split para dividir cada
lista gerando duas novas listas chamadas x e y  importante deixar claro para o usurio
que as duas listas digitadas devem ter a mesma quantidade de nmeros digitados
independente de quantos sejam. Aps isso, faa a tipagem de todos os nmeros das
duas listas como float, ento crie trs novas listas vazias: xquad, yquad e xy. Agora
preencha a lista xquad com cada elemento da lista x elevado ao quadrado, depois faa
o mesmo para yquad, elevando cada elemento de y ao quadrado e por ltimo preencha
a lista xy fazendo a multiplicao de cada valor da lista x pelo seu correspondente na
lista y. No final crie a varivel sx que recebe a soma dos valores da lista x, a varivel sy
que recebe a soma dos valores da lista y, a varivel sxquad, que recebe a soma da lista
xquad, a varivel syquad que recebe a soma da lista de yquad e a varivel sxy que
recebe a soma da lista de xy e no final imprima todas as variveis de soma criadas.

```python
n1 = input('Informe uma sequncia de n nmeros inteiros separados por espaos: ')
n2 = input('''Informe outra sequncia de nmeros inteiros separados por espaos com a mesma quantidade informada na lista anterior''' )
x = n1.split()
y = n2.split()          #len  o tamanho da lista, y tem a mesma quantidade de numeros que x
for i in range(len(x)): #Range(Len(x) no For: Percorra o intervalo (range) e para cada posio (Len) de uma lista (x), faa tal coisa (:)
    x[i] = float(x[i])  #para cada valor de x encontrado na lista x transforme-o em um float de x encontrado nessa lista de x
    y[i] = float(y[i])  #para cada valor de y encontrado na lista y transforme-o em um float de y encontrado nessa lista de y
xquad = []
yquad = []
xy = []
for i in range(len(x)):
    xquad.append(x[i]**2)
    yquad.append(y[i]**2)
    xy.append(x[i]*y[i])
sx = sum(x)
sy = sum(y)
sxquad = sum(xquad)
syquad = sum(yquad)
sxy = sum(xy)
print(f'A soma dos valores de x  {sx}')
print(f'A soma dos valores de y  {sy}')
print(f'A soma dos valores de xquad  {sxquad}')
print(f'A soma dos valores de yquad  {syquad}')
print(f'A soma dos valores de xy  {sxy}')
```

#*Exemplo 4: Faa um script que receba do usurio n nmeros inteiros, armazene os nmeros recebidos a cada momento em que forem recebidos em uma lista e no final exiba a soma desses valores. Considere que o usurio dever informar apenas 1 numero inteiro por caixa de dilogo  permitido usar (While).*

```python
c = 0 # criei um contador para utilizar como parada do lao while
n = int(input('Informe a qtde de nmeros inteiros que deseja na lista: '))
numeros = []
while c < n:
  numeros.append(int(input(f'Informe o {c+1} nmero inteiro da lista: ')))
  c += 1
print(f'A soma dos valores digitados  {sum(numeros)}')
```
